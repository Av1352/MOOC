```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
a=pd.read_csv("/content/FuelConsumption.csv")
a
```

```python
plt.scatter(a["MAKE"],a["MODEL"])
plt.xlabel("Makers",size=20)
plt.ylabel("Models",size=20)
plt.title("Cars",size=40)
plt.rcParams['figure.figsize']=(10,10)
plt.show()
```

```python
n=a["FUELTYPE"]
x=a["FUELCONSUMPTION_CITY"]
# z=a["MODEL"]
y=a["FUELCONSUMPTION_HWY"]
z=a["FUELCONSUMPTION_COMB"]
plt.plot(x,n,"o-.r",mfc="cyan",ms=20,label="Fuel Consumption in City")
plt.plot(y,n,"o-b",mfc="red",ms=20,label="Fuel Consumption in Highway")
plt.plot(z,n,"o:k",mfc="white",ms=20,label="Fuel Consumption in Comb")
plt.ylabel("Fuel type",size=100)
plt.xlabel("Fuel Consumption",size=100)
plt.title("Fuel Type and Consumption in different places",size=150)
plt.xticks(range(20),size=50)
plt.yticks(size=50)
plt.legend(prop={'size':80})
plt.rcParams['figure.figsize']=(90,25)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.bar(a["CYLINDERS"],a["ENGINESIZE"])
plt.xlabel("Cylinders",size=20)
plt.ylabel("Engine Size",size=20)
plt.title("Cylinder Capacity",size=40)


plt.subplot(1,2,2)
plt.bar(a["CO2EMISSIONS"],a["ENGINESIZE"],width=1.5)
plt.xlabel("CO2 Emissions",size=20)
plt.ylabel("Engine Size",size=20)
plt.title("CO2 Emissions",size=40)

plt.rcParams['figure.figsize']=(50,10)
plt.show()
```

```python
plt.pie(a["FUELCONSUMPTION_HWY"],labels=a["MODEL"])
plt.title("Fuel Consumption of Different Cars")
plt.rcParams['figure.figsize']=(13,13)
plt.show()
```
